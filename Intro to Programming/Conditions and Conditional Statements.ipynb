```python
def get_grade(score):
    grade = "A"
    
    print(score)
    
    if score >= 80:
        grade = "B"
    elif score >= 70 & score <= 79:
        grade = "C"
    elif score >= 60 & score <= 69:
        grade = "D"
    elif score < 60:
        grade = "F"
    
    return grade
```

```python
get_grade(0)
```
